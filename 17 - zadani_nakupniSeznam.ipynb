Vytvořte třídu Seznam, která bude mít tyto atributy:

*   Název seznamu
*   Seznam položek
*   Maximální útrata (kolik mám u sebe peněz)

Ta bude obsahovat metody pro get, set, ToString a metody pro přidání a odebrání položky seznamu a zjištění ceny celého nákupu.


```python
class Seznam:
    def __init__(self, nazev, max_utrata, seznam=None):
        if seznam is None:
            seznam = []
        self.seznam = seznam
        self.nazev = nazev
        self.max_utrata = max_utrata

    def seznamToString(self):
        return "".join(polozka.__str__() + "\n" for polozka in self.seznam)

    def _seznam_to_string_old(self):
        text = ""
        for polozka in self.seznam:
            text += polozka.__str__() + "\n"
        return text

    def pridej_polozku(self, polozka):
        if polozka in self.seznam:
            print("Položka již v seznamu je.")
            return

        if self.cena_nakupu() + polozka.cena * polozka.pocet > self.max_utrata:
            print("Nelze přidat – překročen limit útraty.")
            return

        self.seznam.append(polozka)

    def odeber_polozku(self, polozka):
        if polozka in self.seznam:
            self.seznam.remove(polozka)
        else:
            print("Položka v seznamu není.")

    def cena_nakupu(self):
        return sum(polozka.cena * polozka.pocet for polozka in self.seznam)

    def cena_nakupu_old(self):
        cena = 0
        for polozka in self.seznam:
            cena += polozka.cena * polozka.pocet
        return cena





#s1 = Seznam(None,"Můj nákupní seznam")
#s1 = Seznam([]"M")


```

```python
#TODO
class Polozka:
    def __init__(self, nazev, cena, pocet=1):
        self.data = {
            "nazev": nazev,
            "cena": cena,
            "pocet": pocet
        }

    def __str__(self):
        return f"{self.data['nazev']} – {self.data['pocet']} ks × {self.data['cena']} Kč"

    def celkova_cena(self):
        return self.data["cena"] * self.data["pocet"]

    def nastav_pocet(self, novy_pocet):
        if novy_pocet > 0:
            self.data["pocet"] = novy_pocet
        else:
            print("Počet musí být kladné číslo.")

    def nastav_cenu(self, nova_cena):
        if nova_cena >= 0:
            self.data["cena"] = nova_cena
        else:
            print("Cena nemůže být záporná.")

    def vrat_hodnotu(self, klic):
        return self.data.get(klic, None)

    def nastav_hodnotu(self, klic, hodnota):
        self.data[klic] = hodnota

p1 = Polozka("Jablka", 12, 5)

print(p1)  


print(p1.celkova_cena())


p1.nastav_pocet(10)
print(p1.celkova_cena())


p1.nastav_hodnotu("poznamka", "bio kvalita")
print(p1.vrat_hodnotu("poznamka"))





```
