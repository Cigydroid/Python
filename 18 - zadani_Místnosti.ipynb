1. název

**Vytvořte metody:**


1.   addExit, getNazev, getExity, getPoklad, getNepritel, toString

```python
#TODO
class Room:
    def __init__(self,nazev):
        self.nazev = nazev
        self.seznam_exitu = []
        self.poklad = ""
        self.nepritel = ""
 
class Room:
    def __init__(self, nazev):
        self.nazev = nazev
        self.seznam_exitu = []
        self.poklad = ""
        self.nepritel = ""


    def addExit(self, exit):
        self.seznam_exitu.append(exit)

 
    def getNazev(self):
        return self.nazev

    def setNazev(self, nazev):
        self.nazev = nazev

    def getExity(self):
        return self.seznam_exitu

    def getPoklad(self):
        return self.poklad

    def getNepritel(self):
        return self.nepritel

    def toString(self):
        return (
            f"room: {self.nazev}\n"
            f"vexity: {self.seznam_exitu}\n"
            f"Poklad: {self.poklad}\n"
            f"nepritel: {self.nepritel}"
        )


```

**Zde je vytvořená jednoduchá mapa s výpisy**

```python
room1 = Room("Vstup do jeskyně")
room2 = Room("Jezírko")
room3 = Room("Předsíň draka")
room4 = Room("Dračí slůj")
room5 = Room("Vyhlídka")

room1.addExit(room5)
room5.addExit(room1)
room1.addExit(room2)
room1.addExit(room4)
room2.addExit(room1)
room4.addExit(room1)
room4.addExit(room3)
room4.addExit(room5)
room3.addExit(room4)
room5.addExit(room4)

aktualni = room1
print(aktualni.getSeznamExitu())
print(aktualni.getNepritel())
print(aktualni.getPoklad())
print(aktualni)

#chci se přesunout tak si změním aktualní na jeden z východu
aktualni = aktualni.seznamExitu[0]
print(aktualni)
#aktualni = aktualni.seznamExitu[2] # již nemá 3 exit --> skončí chybou
#print(aktualni)
aktualni = aktualni.seznamExitu[1]
print(aktualni)
```
