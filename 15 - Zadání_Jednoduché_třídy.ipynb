1.   název:str

````python
class Pojistovna:
    def __init__(self,nazev,zkratka,sídlo):
        self.nazev = nazev
        if len(zkratka) <=3:
            self.zkratka = zkratka
        else:
            print("zkratkaj e moc dlouhá vznike zkrácený tvar ")
            self.zkratka = zkratka[0:3]
        self.sídlo = sídlo

    def get_nazev(self):
        return self.nazev
    def set_nazev(self,nove_nazev):
        self.nazev = nove_nazev
        print("nazev byl změněn")

    def get_zkratka(self):
        return self.zkratka
    def set_zkratka(self,nove_zkratka):
        self.zkratka = nove_zkratka
        print("zkratka byla změněna")
  
    def get_sídlo(self):
        return self.sídlo
    def set_sídlo(self,nove_sídlo):
        self.sídlo = nove_sídlo
        print("sídlo bylo změněno")

    def __str__(self):
        return f"""
        Název: {self.nazev} 
        Zkratka: {self.zkratka}
        Sídlo: {self.sídlo}
"""
pojistovna1 = Pojistovna("Zelená","ZLN","Praha 4")

print(pojistovna1)
pojistovna1.set_nazev("Modrá")
pojistovna1.set_zkratka("MDR")
pojistovna1.set_sídlo("Brno 6")
print(pojistovna1)    
````

# Počasí

Vytvořte třídu Pocasi, která bude mít atributy:

1.   vítr:float
2.   povetrnostniPodminky:str
3.   teplota:float
4.   vlhkost:float

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše tomto počasí


````python
class Pocasi:
    def __init__(self,vítr,povetrnostnipodminky,teplota,vlhkost):
        self.vítr = vítr
        self.povetrnostnipodminky = povetrnostnipodminky
        self.teplota = teplota
        self.vlhkost = vlhkost


    def get_vítr(self):
        return self.vítr
    def set_vítr(self,nove_vítr):
        self.vítr = nove_vítr
        print("vítr byl změněn")
    
    def get_povetrnostnipodminky(self):
        return self.povetrnostnipodminky
    def set_povetrnostnipodminky(self,nove_povetrnostnipodminky):
        self.povetrnostnipodminky = nove_povetrnostnipodminky
        print("povetrnostnipodminky byly změněny")

    def get_teplota(self):
        return self.teplota
    def set_teplota(self,nove_teplota):
        self.teplota = nove_teplota
        print("teplota byla změněna")

    def get_vlhkost(self):
        return self.vlhkost
    def set_vlhkost(self,nove_vlhkost):
        self.vlhkost = nove_vlhkost
        print("vlhkost byla změněna")
    
    def __str__(self):
        return f"""
        Vítr: {self.vítr}
        Povetrnostnipodminky: {self.povetrnostnipodminky} 
        Teplota: {self.teplota}
        Vlhkost: {self.vlhkost}
"""
    
pocasi1 = Pocasi("km/h","dobré","50°","6%")
print(pocasi1)
pocasi1.set_vítr("km/h")
pocasi1.set_povetrnostnipodminky("špatné")
pocasi1.set_teplota("-20°C")
pocasi1.set_vlhkost("16%")
print(pocasi1)


````
